```python
def sentiment(title, description):
    chat_completion = client.chat.completions.create(
    messages=[
        {
            "role": "system",
            "content": "You will have to do sentiment analysis on News title and description and provide the sentiment of the news.",
        },
        {
            "role": "user",
            "content": f"Title: {title}\nDescription: {description}",
        }
    ],
    model="llama-guard-3-8b",
    temperature=0.5,
    max_completion_tokens=1024,
    top_p=1,
    stop=None,
    stream=False,
)
    print(chat_completion.choices[0].message.content.strip())
```

```python
def sentiment(title, description):
    chat_completion = client.chat.completions.create(
        messages=[
            {
                "role": "system",
                "content": "You will analyze the sentiment of a news title and description. Respond with 'positive', 'negative', or 'neutral'.",
            },
            {
                "role": "user",
                "content": f"Title: {title}\nDescription: {description}",
            }
        ],
        model="llama3-70b-8192",  # Replace with a general-purpose model
        temperature=0.5,
        max_tokens=50,
        top_p=1,
    )
    
    return chat_completion.choices[0].message.content.strip()

# Example usage
result = sentiment(response['results'][0]['title'], response['results'][0]['description'])
print(result)
```
